1. Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         
         

```
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
```

​​​​​

````python
import matplotlib.pyplot as plt
Days=[1,2,3,4,5,6,7,8]
Speed=[60,62,61,58,56,57,46,63]
plt.plot(Days, Speed, 'r', linewidth=3)
plt.xlabel("Days")
plt.ylabel("Speed")
plt.title("Car speed measurement")
plt.show()
````

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

````python
import numpy as np
t = np.linspace(0,4*np.pi,100)
sint = np.sin(t)
cost = np.cos(t)

plt.figure(figsize=(12,4))
plt.plot(t,sint,label="sint",linewidth=3)
plt.plot(t,cost,label="cost",linewidth=3)
plt.legend()
plt.xlabel("Time t")
plt.ylabel("Magnitude")
plt.title("Superimposition sine/cosine curve", fontsize=20)
````

5. Plot Simple bar chart showing popularity of Programming Languages.
```
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
```
Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

````python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = np.array([56, 39, 34, 34, 29])
Security = np.array([44 ,36 ,55, 50, 42])

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(12,12))

ax1.bar(Languages, Popularity, color="b", edgecolor="r")
ax1.set_title("Bar plot of popularity",fontsize=20)
ax1.set_xlabel("Programming language")
ax1.set_ylabel("Popularity")


ax2.hist(Security, color="g", edgecolor="r")
ax2.set_title("Histogram for security",fontsize=20)
ax2.set_xlabel("Bins")
ax2.set_ylabel("Security")

ax3.barh(Languages, Security, color="y", edgecolor="r")
ax3.set_title("Horizontal bar plot of security",fontsize=20)
ax3.set_xlabel("Programming Languages")
ax3.set_ylabel("Security")

# fig.suptitle("Analysis of programming languages",fontsize=25)
fig.tight_layout()

````

6. Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.
  ```
  Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]
  ```

  Histogram showing Below Average, Average and Execellent distribution

  40-60: Below Average

  60-80: Average

  80-100: Excellent

````python
Marks = np.array([ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])
plt.hist(Marks,color='r',edgecolor='b')
plt.vlines([40,60,80,100],ymin=0,ymax=4)
plt.xlabel("Bins")
plt.ylabel("Marks")
plt.title("Histogram of Marks")
plt.show()
print("The number of students with ____ marks")
print("1. Below Average:",len(Marks[np.logical_and(Marks>=40,Marks<60)]))
print("2. Average:",len(Marks[np.logical_and(Marks>=60,Marks<80)]))
print("3. Excellent:",len(Marks[np.logical_and(Marks>=80,Marks<100)]))
````

````python
bins = [40,60,80,100]
plt.hist(Marks,bins,color='r',edgecolor='b')
plt.xlabel("Bins")
plt.ylabel("Marks")
plt.title("Histogram of Marks")
plt.show()
````
